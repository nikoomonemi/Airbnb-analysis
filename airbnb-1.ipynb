```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt

```

```python
list = pd.read_csv("/Users/nikoo/Documents/analys project/airbnb/AB_NYC_2019.csv")
list.head()
```

```python
list.info()
```

```python
list.describe()
```

```python
#cleaning data
list = (list[list["price"]<1000]) 
list = list[list["price"]>0]
```

```python
#location factor
location_factor = list.groupby("neighbourhood_group")["price"].mean().sort_values(ascending=False)
location_factor
```

```python
#visualization of neighbourhood factor on price
neighbourhood_chart = sns.barplot(data = list, x = "neighbourhood_group", y = "price", color = "red")
plt.title("neighbourhood on price")
plt.xlabel("neighbourhood_group")
plt.ylabel("price")
plt.show()
```

```python
#roomtype factor
plt.figure(figsize=(10,10))
roomtype_afect = sns.barplot(data = list, x="room_type", y= "price", color = "brown")
plt.title("price on room type")
plt.xlabel("room type")
plt.ylabel("price")
plt.show()
```

```python
# count of reviwe factor
plt.figure(figsize=(10,10))
reviwe_factor = sns.scatterplot(data= list, x= "number_of_reviews", y = "price", color = "orange")
plt.title("number of reviwes on price")
plt.xlabel("number of reviwes")
plt.ylabel("price")
plt.show()
```

```python
#correlation heatmap.
corr= list[["price", "minimum_nights","number_of_reviews","availability_365"]].corr()
corr
```

```python
corr_chart = sns.heatmap(corr, annot= True, cmap= "coolwarm")
```
